# FAO Cost and Affordability of Healthy Diet (CoAHD) - Pakistan EDA



## Exploratory Data Analysis using Plotly Visualizations

**Dataset Source:** [FAO FAOSTAT](https://www.fao.org/faostat/en/#data/CAHD)  
**Country:** Pakistan  
**Time Period:** 2017-2024  
**Last Updated:** January 2026

This notebook provides a comprehensive exploratory data analysis of Pakistan's healthy diet affordability metrics, including costs, unaffordability prevalence, and food category breakdowns.




## Section 1: Import Required Libraries

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

print("Libraries imported successfully!")
```

## Section 2: Load and Explore the Dataset

```python
# Load the dataset
df = pd.read_csv('FAOSTAT_data_2017-2026.csv')

# Display basic information
print("=" * 80)
print("DATASET OVERVIEW")
print("=" * 80)
print(f"\nDataset Shape: {df.shape[0]} rows  {df.shape[1]} columns")
print(f"\nColumn Names and Data Types:")
print(df.dtypes)
print(f"\nFirst Few Rows:")
df.head(10)
```

```python
# Check for missing values
print("\n" + "=" * 80)
print("MISSING VALUES ANALYSIS")
print("=" * 80)
missing_data = df.isnull().sum()
print(missing_data[missing_data > 0] if missing_data.sum() > 0 else "No missing values found!")

# Summary statistics for numerical columns
print("\n" + "=" * 80)
print("SUMMARY STATISTICS")
print("=" * 80)
print(df.describe())

# Unique values in key columns
print("\n" + "=" * 80)
print("UNIQUE VALUES IN KEY COLUMNS")
print("=" * 80)
print(f"Unique Items: {df['Item'].nunique()}")
print(f"Unique Years: {df['Year'].nunique()}")
print(f"Unique Units: {df['Unit'].nunique()}")
print(f"\nItem Categories:")
print(df['Item'].unique())
```

## Section 3: Data Cleaning and Preparation

```python
# Create a copy for processing
df_clean = df.copy()

# Convert Year to integer
df_clean['Year'] = pd.to_numeric(df_clean['Year'], errors='coerce')

# Convert Value to numeric
df_clean['Value'] = pd.to_numeric(df_clean['Value'], errors='coerce')

# Remove any duplicate rows
df_clean = df_clean.drop_duplicates()

# Display cleaning summary
print("=" * 80)
print("DATA CLEANING SUMMARY")
print("=" * 80)
print(f"Original shape: {df.shape}")
print(f"Cleaned shape: {df_clean.shape}")
print(f"Duplicates removed: {df.shape[0] - df_clean.shape[0]}")
print(f"\nData types after cleaning:")
print(df_clean[['Year', 'Value']].dtypes)
print(f"\nData preparation completed successfully!")

# Show cleaned data sample
print("\nSample of cleaned data:")
df_clean.head()
```

## Section 4: Cost of Healthy Diet Trends Analysis

### Cost of Healthy Diet (CoHD) - Main Indicator

```python
# Filter data for CoHD metrics
cohd_lcu = df_clean[df_clean['Item'].str.contains('LCU per person per day', case=False, na=False) & 
                      df_clean['Item'].str.contains('Cost of a healthy diet', case=False, na=False)].copy()
cohd_ppp = df_clean[df_clean['Item'].str.contains('PPP dollar per person per day', case=False, na=False) & 
                     df_clean['Item'].str.contains('Cost of a healthy diet', case=False, na=False)].copy()

# Sort by year
cohd_lcu = cohd_lcu.sort_values('Year')
cohd_ppp = cohd_ppp.sort_values('Year')

# Create interactive line chart for CoHD trends (LCU and PPP)
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add LCU trace
fig.add_trace(
    go.Scatter(x=cohd_lcu['Year'], y=cohd_lcu['Value'], 
               mode='lines+markers', name='LCU per capita per day',
               line=dict(color='#1f77b4', width=3),
               marker=dict(size=8), hovertemplate='<b>Year:</b> %{x}<br><b>Cost (LCU):</b> %{y:.2f}<extra></extra>'),
    secondary_y=False
)

# Add PPP trace
fig.add_trace(
    go.Scatter(x=cohd_ppp['Year'], y=cohd_ppp['Value'], 
               mode='lines+markers', name='Int$ (PPP) per capita per day',
               line=dict(color='#ff7f0e', width=3),
               marker=dict(size=8), hovertemplate='<b>Year:</b> %{x}<br><b>Cost (PPP$):</b> %{y:.2f}<extra></extra>'),
    secondary_y=True
)

# Update axes labels
fig.update_yaxes(title_text="<b>Cost (LCU/cap/d)</b>", secondary_y=False)
fig.update_yaxes(title_text="<b>Cost (Int$ PPP/cap/d)</b>", secondary_y=True)
fig.update_xaxes(title_text="<b>Year</b>")

fig.update_layout(
    title="<b>Cost of Healthy Diet Trends in Pakistan (2017-2024)</b>",
    hovermode='x unified',
    height=500,
    template='plotly_white',
    font=dict(size=12)
)

fig.show(renderer='iframe')

print("Cost of Healthy Diet Trends:")
print(f"LCU 2017: {cohd_lcu[cohd_lcu['Year']==2017]['Value'].values[0]:.2f}  2024: {cohd_lcu[cohd_lcu['Year']==2024]['Value'].values[0]:.2f}")
print(f"PPP 2017: {cohd_ppp[cohd_ppp['Year']==2017]['Value'].values[0]:.2f}  2024: {cohd_ppp[cohd_ppp['Year']==2024]['Value'].values[0]:.2f}")
print(f"LCU Growth: {((cohd_lcu[cohd_lcu['Year']==2024]['Value'].values[0] / cohd_lcu[cohd_lcu['Year']==2017]['Value'].values[0]) - 1) * 100:.1f}%")
```

## Section 5: Affordability Metrics Visualization

### Prevalence of Unaffordability (PUA) and Number Unable to Afford (NUA)

```python
# Filter PUA and NUA data
pua_data = df_clean[df_clean['Item'].str.contains('Prevalence of unaffordability', case=False, na=False)].sort_values('Year')
nua_data = df_clean[df_clean['Item'].str.contains('Number of people unable', case=False, na=False)].sort_values('Year')

# Create subplots
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=("Prevalence of Unaffordability (PUA)", "Number Unable to Afford (NUA)"),
    specs=[[{"secondary_y": False}, {"secondary_y": False}]]
)

# Add PUA bar chart
fig.add_trace(
    go.Bar(x=pua_data['Year'], y=pua_data['Value'], 
           name='PUA (%)', marker_color='#d62728',
           hovertemplate='<b>Year:</b> %{x}<br><b>Unaffordability:</b> %{y:.1f}%<extra></extra>'),
    row=1, col=1
)

# Add NUA bar chart
fig.add_trace(
    go.Bar(x=nua_data['Year'], y=nua_data['Value'], 
           name='NUA (millions)', marker_color='#2ca02c',
           hovertemplate='<b>Year:</b> %{x}<br><b>Unable to Afford:</b> %{y:.1f}M<extra></extra>'),
    row=1, col=2
)

# Update layout
fig.update_xaxes(title_text="Year", row=1, col=1)
fig.update_xaxes(title_text="Year", row=1, col=2)
fig.update_yaxes(title_text="Percentage (%)", row=1, col=1)
fig.update_yaxes(title_text="Population (Millions)", row=1, col=2)

fig.update_layout(
    title_text="<b>Diet Affordability Crisis in Pakistan</b>",
    height=500,
    showlegend=False,
    template='plotly_white',
    font=dict(size=11),
    hovermode='x unified'
)

fig.show()

# Print key statistics
print("=" * 80)
print("AFFORDABILITY METRICS SUMMARY")
print("=" * 80)
print("\nPrevalence of Unaffordability (PUA) - % of population:")
print(pua_data[['Year', 'Value']].to_string(index=False))
print(f"\nAverage PUA: {pua_data['Value'].mean():.1f}%")
print(f"Maximum PUA: {pua_data['Value'].max():.1f}% (Year {pua_data[pua_data['Value']==pua_data['Value'].max()]['Year'].values[0]:.0f})")
print(f"Minimum PUA: {pua_data['Value'].min():.1f}% (Year {pua_data[pua_data['Value']==pua_data['Value'].min()]['Year'].values[0]:.0f})")

print("\n" + "-" * 80)
print("\nNumber Unable to Afford (NUA) - Millions:")
print(nua_data[['Year', 'Value']].to_string(index=False))
print(f"\nAverage NUA: {nua_data['Value'].mean():.1f}M")
print(f"Maximum NUA: {nua_data['Value'].max():.1f}M (Year {nua_data[nua_data['Value']==nua_data['Value'].max()]['Year'].values[0]:.0f})")
print(f"Minimum NUA: {nua_data['Value'].min():.1f}M (Year {nua_data[nua_data['Value']==nua_data['Value'].min()]['Year'].values[0]:.0f})")
```

## Section 6: Food Category Cost Breakdown

### Cost Analysis by Food Components (2021 Data)

```python
# Filter food categories data for 2021
categories_2021 = df_clean[(df_clean['Year'] == 2021) & 
                           (df_clean['Item'].str.contains('Cost of', case=False, na=False)) &
                           (~df_clean['Item'].str.contains('Cost of a healthy diet', case=False, na=False))].copy()

# Separate LCU and PPP data
categories_lcu = categories_2021[categories_2021['Unit'].str.contains('LCU', case=False, na=False)].sort_values('Value', ascending=False)
categories_ppp = categories_2021[categories_2021['Unit'].str.contains('PPP', case=False, na=False)].sort_values('Value', ascending=False)

# Extract category names for labeling
def extract_category(item_str):
    if 'starchy staples' in item_str.lower():
        return 'Starchy Staples'
    elif 'animal source' in item_str.lower():
        return 'Animal Source Foods'
    elif 'legumes' in item_str.lower():
        return 'Legumes, Nuts & Seeds'
    elif 'vegetables' in item_str.lower():
        return 'Vegetables'
    elif 'fruits' in item_str.lower():
        return 'Fruits'
    elif 'oils and fats' in item_str.lower():
        return 'Oils and Fats'
    return 'Other'

categories_lcu['Category'] = categories_lcu['Item'].apply(extract_category)
categories_ppp['Category'] = categories_ppp['Item'].apply(extract_category)

# Create subplots
fig = make_subplots(
    rows=1, cols=2,
    specs=[[{"type":"pie"}, {"type":"bar"}]],
    subplot_titles=("Cost Distribution by Food Category (LCU)", "Cost Comparison (PPP)")
)

# Pie chart for LCU
fig.add_trace(
    go.Pie(labels=categories_lcu['Category'], values=categories_lcu['Value'],
            name='LCU', hovertemplate='<b>%{label}</b><br>Cost: %{value:.2f} LCU<br>Percentage: %{percent}<extra></extra>'),
    row=1, col=1
)

# Bar chart for PPP
fig.add_trace(
    go.Bar(x=categories_ppp['Category'], y=categories_ppp['Value'],
           marker_color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b'],
           name='PPP', hovertemplate='<b>%{x}</b><br>Cost: %{y:.2f} Int$<extra></extra>'),
    row=1, col=2
)

fig.update_layout(
    title_text="<b>Healthy Diet Cost Breakdown by Food Components (2021)</b>",
    height=500,
    showlegend=False,
    template='plotly_white',
    font=dict(size=11),
    hovermode='closest'
)

fig.update_xaxes(title_text="Food Category", row=1, col=2)
fig.update_yaxes(title_text="Cost (Int$ PPP)", row=1, col=2)

fig.show(renderer='iframe')

# Print detailed breakdown
print("=" * 80)
print("FOOD CATEGORY COST BREAKDOWN (2021)")
print("=" * 80)
print("\nCost in Local Currency Units (LCU/cap/day):")
print(categories_lcu[['Category', 'Value']].to_string(index=False))
print(f"\nTotal: {categories_lcu['Value'].sum():.2f} LCU")

print("\n" + "-" * 80)
print("\nCost in PPP Dollars (Int$/cap/day):")
print(categories_ppp[['Category', 'Value']].to_string(index=False))
print(f"\nTotal: {categories_ppp['Value'].sum():.2f} Int$")

# Calculate percentages
print("\n" + "-" * 80)
print("\nPercentage Distribution:")
for idx, row in categories_lcu.iterrows():
    pct = (row['Value'] / categories_lcu['Value'].sum()) * 100
    print(f"{row['Category']:25s}: {pct:6.1f}%")
```

## Section 7: Time Series Analysis and Insights

```python
# Year-over-year growth analysis
print("=" * 80)
print("YEAR-OVER-YEAR GROWTH ANALYSIS")
print("=" * 80)

# Calculate growth for CoHD LCU
cohd_lcu_growth = cohd_lcu[['Year', 'Value']].sort_values('Year').reset_index(drop=True)
cohd_lcu_growth['YoY_Growth_%'] = cohd_lcu_growth['Value'].pct_change() * 100

print("\nCost of Healthy Diet (LCU) - Year-over-Year Growth:")
print(cohd_lcu_growth.to_string(index=False))

# Calculate annual averages
print("\n" + "=" * 80)
print("ANNUAL AVERAGES BY INDICATOR")
print("=" * 80)

indicators = {
    'Cost of Healthy Diet (LCU)': cohd_lcu,
    'Cost of Healthy Diet (PPP)': cohd_ppp,
    'Prevalence of Unaffordability': pua_data,
    'Number Unable to Afford': nua_data
}

for indicator_name, data in indicators.items():
    avg_value = data['Value'].mean()
    print(f"\n{indicator_name}:")
    print(f"  Average: {avg_value:.2f}")
    print(f"  Min: {data['Value'].min():.2f} (Year {data.loc[data['Value'].idxmin(), 'Year']:.0f})")
    print(f"  Max: {data['Value'].max():.2f} (Year {data.loc[data['Value'].idxmax(), 'Year']:.0f})")
    print(f"  Std Dev: {data['Value'].std():.2f}")

# Create a comprehensive summary visualization
fig = go.Figure()

# Normalize values for comparison (0-100 scale)
cohd_lcu_norm = (cohd_lcu.sort_values('Year')['Value'] - cohd_lcu['Value'].min()) / (cohd_lcu['Value'].max() - cohd_lcu['Value'].min()) * 100
pua_norm = (pua_data.sort_values('Year')['Value'] - pua_data['Value'].min()) / (pua_data['Value'].max() - pua_data['Value'].min()) * 100

fig.add_trace(go.Scatter(
    x=cohd_lcu.sort_values('Year')['Year'],
    y=cohd_lcu_norm,
    mode='lines+markers',
    name='CoHD Cost Trend (LCU)',
    line=dict(color='#1f77b4', width=3)
))

fig.add_trace(go.Scatter(
    x=pua_data.sort_values('Year')['Year'],
    y=pua_norm,
    mode='lines+markers',
    name='Affordability Crisis Trend',
    line=dict(color='#d62728', width=3)
))

fig.update_layout(
    title="<b>Normalized Trends Comparison (2017-2024)</b>",
    xaxis_title="Year",
    yaxis_title="Normalized Index (0-100)",
    height=500,
    template='plotly_white',
    hovermode='x unified',
    font=dict(size=12)
)

fig.show(renderer='iframe')

print("\n" + "=" * 80)
print("KEY INSIGHTS")
print("=" * 80)
print(f"""
1. COST ESCALATION:
   - Healthy diet cost increased by {((cohd_lcu[cohd_lcu['Year']==2024]['Value'].values[0] / cohd_lcu[cohd_lcu['Year']==2017]['Value'].values[0]) - 1) * 100:.1f}% in LCU from 2017-2024
   - This represents a significant inflationary pressure on food accessibility

2. AFFORDABILITY CRISIS:
   - On average, {pua_data['Value'].mean():.1f}% of Pakistan's population cannot afford a healthy diet
   - Over {nua_data['Value'].mean():.1f} million people are unable to afford nutritious food annually

3. FOOD CATEGORY INSIGHTS (2021):
   - Animal source foods account for {(categories_lcu[categories_lcu['Category']=='Animal Source Foods']['Value'].values[0] / categories_lcu['Value'].sum()) * 100:.1f}% of healthy diet cost
   - Fruits and vegetables together comprise {(categories_lcu[categories_lcu['Category'].isin(['Vegetables', 'Fruits'])]['Value'].sum() / categories_lcu['Value'].sum()) * 100:.1f}% of total cost

4. RECENT TRENDS:
   - Peak unaffordability was {pua_data['Value'].max():.1f}% in {pua_data[pua_data['Value']==pua_data['Value'].max()]['Year'].values[0]:.0f}
   - Peak unable to afford: {nua_data['Value'].max():.1f}M people in {nua_data[nua_data['Value']==nua_data['Value'].max()]['Year'].values[0]:.0f}

5. POLICY IMPLICATIONS:
   - The persistent high levels of diet unaffordability suggest urgent need for:
     * Food security interventions
     * Income support programs
     * Agricultural productivity improvements
     * Price stabilization measures
""")
```
